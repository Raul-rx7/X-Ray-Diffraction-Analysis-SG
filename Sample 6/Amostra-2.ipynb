```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns, cufflinks as cf, plotly.express as px
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

%matplotlib inline
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
a2 = pd.read_csv('Amostra 6.csv', names=['x', 'y'])
a2 = a2[['x', 'y']].dropna()
a2.head()
```

## Plotando o Difratograma com a biblioteca matplotlib:

```python
plt.figure(figsize=(10, 4), clear=True)
plt.plot(a2['x'], a2['y'], color='blue', linewidth=0.2)
plt.xlabel(r'2$\theta$(graus)')
plt.ylabel('Intensidade(CPS)')
plt.legend(['Amostra 2'], loc=1)
plt.title('DRX');
```

## Plotando com a biblioteca Plotly:

É possível plotar um gráfico interativo e verificar os pontos cruciais para a análise do material em questão.

```python
px.line(a2, x="x", y="y", title='DRX: Amostra 2', labels={'x': 'Graus(°) ', 'y':'Intensidade(CPS)'})
```
