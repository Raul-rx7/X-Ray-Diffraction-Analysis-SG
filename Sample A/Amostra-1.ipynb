```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns, cufflinks as cf, plotly.express as px
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

%matplotlib inline
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
a1 = pd.read_fwf('Amostra A.txt')
a1.to_csv('Amostra A.csv')
```

```python
a1 = pd.read_csv('Amostra A.csv', names=['x', 'y'])
```

```python
a1 = a1[['x', 'y']].dropna()
a1.head()
```

## Plotando Difratograma com a biblioteca matplotlib:

```python
plt.figure(figsize=(10, 4), clear=True)
plt.plot(a1['x'], a1['y'], color='blue', linewidth=0.2)
plt.xlabel(r'2$\theta$(graus)')
plt.ylabel('Intensidade(CPS)')
plt.legend(['Amostra 1'], loc=1)
plt.title('DRX');
```

## Plotando com a biblioteca Plotly:

É possível plotar um gráfico interativo e verificar os pontos cruciais para a análise do material em questão.

```python
fig = px.line(a1, x="x", y="y", title='DRX: Amostra 1', 
        labels={'x': 'Graus(°) ', 'y':'Intensidade(CPS)'})
fig.update_layout(
    autosize=False,
    width=800,
    height=500)
fig.show();
```
